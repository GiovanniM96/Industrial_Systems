```python
import pandas as pd

# leggi il file excel
df = pd.read_excel("MATRICE DEI FLUSSI CARRELLI ELEVATORI.xlsx")

df.head(50)
```

## From Matplotlib

```python
import matplotlib.pyplot as plt

# genera la tabella
table = plt.table(cellText=df.head(22).values, colLabels=df.columns, loc='center')

# nascondi gli assi
plt.axis('off')

# salva la tabella come immagine
plt.savefig('table.png')
```

# Export customized table 

```python
import pandas as pd

# leggi il file excel
df = pd.read_excel("Progetto impianti.xlsx", sheet_name="BARICENTRI")

# personalizzare l'estetica della tabella
style = """
<style>
table {
  font-family: Arial, Helvetica, sans-serif;
  border-collapse: collapse;
  width: 100%;
}
th, td {
  border: 1px solid #ddd;
  text-align: left;
  padding: 8px;
}
tr:nth-child(even){
  background-color: #f2f2f2;
}
th {
  background-color: #f7943e;
  color: white;
  font-weight: bold;
}
</style>

"""

# esportare il dataframe in formato HTML con lo stile personalizzato
html = style + df.to_html(index=False)

# scrivere il contenuto HTML su un file
with open('BARICENTRI.html', 'w') as f:
    f.write(html)
```
